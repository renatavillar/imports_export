```python
data2 = copy.deepcopy(data)
data2['date'] = pd.to_datetime(data2['date'], format = '%Y-%m-%d')
data2 = data2[(data2['date'] >= '2016-12-01') & ((data2['product'] == 'corn') | 
                                                (data2['product'] == 'sugar'))]
print(data2[data2['country'] == 'Brazil'])
data3 = copy.deepcopy(data)
data3['date'] = pd.to_datetime(data3['date'], format = '%Y-%m-%d')
data3['country'] = data3['country'].astype(str)
data3['product'] = data3['product'].astype(str)
data3['type'] = data3['type'].astype(str)
data3 = data3[(data3['product'] == 'corn') | (data3['product'] == 'sugar')]
data3 = data3[(data3['date'] >= '2016-12-01')]
print('----------------------------------------')
agg_product = data3.groupby(['product', 'country']).agg({'tons': sum, 'usd': sum})
g = agg_product['tons'].groupby(level=0, group_keys=False)
h = agg_product['usd'].groupby(level=0, group_keys=False)
print('----------------------------------------')
print(g.nlargest(10))
print('----------------------------------------')
print(h.nlargest(10))
```

We can see in the penultimate table above that the 10 most important trade partners for Brazil in terms of 'tons' of 'corn' have been Iran, Japan, Vietnam, Egypt, Spain, South Korea, Taiwan, Malaysia, Bangladesh and Paraguay.

The 10 most important trade partners for Brazil in terms of 'usd' of 'sugar' have been Algeria, Bangladesh, India, United Arab Emirates, Saudi Arabia, Nigeria, Iraq, Egypt, Malausia and Morocco.

In the following, we are going to visualize, through graphics, these results.

## Preparing data for visualizations

```python
sort_tons = g.nlargest(10)
sort_usd = h.nlargest(10)

sort_tons['corn'].keys()
```

```python
perc_corn_tons = sort_tons['corn'].values/sort_tons['corn'].values.sum()
labels_corn_tons = list(sort_tons['corn'].keys())
  
perc_corn_usd = sort_usd['corn'].values/sort_usd['corn'].values.sum()
labels_corn_usd = list(sort_usd['corn'].keys())
    
perc_sugar_tons = sort_tons['sugar'].values/sort_tons['sugar'].values.sum()
labels_sugar_tons = list(sort_tons['sugar'].keys())
                                 
perc_sugar_usd = sort_usd['sugar'].values/sort_usd['sugar'].values.sum()
labels_sugar_usd = list(sort_usd['sugar'].keys())
```

```python
gs = gridspec.GridSpec(nrows=2, ncols=1)
pl.figure()
plt.figure(figsize=(12,13))

plt.style.use('ggplot')
plt.figure(figsize=(12,13))
ax1 = pl.subplot(gs[0, 0])
ax1.pie(perc_corn_tons, labels=labels_corn_tons, autopct='%1.1f%%',
        shadow=True, startangle=400)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('corn (tons)')
plt.show()

plt.style.use('ggplot')
plt.figure(figsize=(12,13))
ax2 = pl.subplot(gs[1, 0])
ax2.pie(perc_corn_usd, labels=labels_corn_usd, autopct='%1.1f%%',
        shadow=True, startangle=400)
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('corn (usd)')
plt.show()
```

In the pie charts above, we can see the percentual of participation of the 10 most important trade partners for Brazil in terms of 'tons' of 'corn': As we already seen, these traders are: Iran(22,42% of total of 'tons'), Japan(13%), Vietnam(12,8%) and so on.

# Question 6
Now, we ask you to show your modelling skills. Feel free to use any type of modelling approach, but bear in mind that the modelling approach depends on the nature of your data, and so different models yield different estimates and forecasts. To help you out in this task we also provide you with a dataset of possible covariates (.xlsx). They all come from public sources (IMF, World Bank) and are presented in index number format. Question: What should be the total brazilian soybeans, soybean_meal, and corn export forecasts, in tons, for the next 11 years (2020-2030)? We’re mostly interested in the annual forecast.

I tested three different models: Simple and Multiple Linear Regression, Polynomial Regression, Random Forest and XGBoost. In neither case I got a really good fit. The best model is XGBoost. With 500000 trees (n_estimates parameter equals to 500000), I got an accuracy of around -0.14, which is very bad. I made a features selection, testing different combinations of features, but without a really good fit. It looks like if we increase significantly the number of trees in the XGBoost algorithm (n_estimators much more than 60000), we could get a good fit, but It would be very expensive computationally.

```python
df = pd.read_excel('covariates.xlsx')
df.isna().sum()
```

```python
df['price_soybeans'].fillna(df['price_soybeans'].mean(), inplace=True)
df['price_corn'].fillna(df['price_corn'].mean(), inplace=True)
df['price_soybean_meal'].fillna(df['price_soybean_meal'].mean(), inplace=True)
df['gpd_netherlands'].fillna(df['gpd_netherlands'].mean(), inplace=True)
df.tail()
```

```python
comex = pd.read_csv('data_comexstat.csv', sep=',', encoding='cp860')
for i in range(0, len(comex)):
    comex['date'][i] = int(comex['date'][i][0:4])
```

```python
comex = comex[comex['type'] == 'Export']
```

```python
covariates = copy.deepcopy(df)
covariates = pd.DataFrame(df)
covariates['year'] = covariates['year'].astype(int)
covariates = covariates[covariates['year'] >= 1997]
covariates = covariates[covariates['year'] < 2020]
covariates.rename(columns={'year':'date'}, inplace=True)
covariates.set_index('date', inplace=True)
merged = pd.merge(covariates, comex, on='date')
merged.drop(['state', 'route', 'usd', 'country'], axis=1, inplace=True)
```

```python
merged.dtypes
```

```python
soybeans = merged[merged['product'] == 'soybeans']
soybeans.drop(['product', 'type'], axis=1, inplace=True)
l = len(soybeans)
L = int(0.8*l)
X_train_soybeans = soybeans.iloc[0:L, 1:14]
y_train_soybeans = soybeans.iloc[0:L, 13]
```

## Correlation matrix

```python
corr_soybeans = X_train_soybeans.corr()
ax = plt.subplots(figsize=(11, 8))
sns.heatmap(corr_soybeans,  annot=True, annot_kws={"size": 10})
```

There are some very correlated features. For example: 'gdp_china' with 'gdp_egypt', 'gdp_china' with 'gdp_world', 'gdp_china' with 'gdp_thailand', 'gdp_iran' with 'gdp_netherland'. In these cases, we consider only one feature in the highly correlated pair. For example, we can consider 'gdp_china' and discard 'gdp_egypt'. Because of the high correlation, it suggests that they have the same power in the model. It is redundant to consider both and, moreover, could worse the performance of the model. But we need to take care with high correlations. For example, 'price_corn' is highly correlated with 'price_soybeans', but it is not clear, with the dataset we have in hands, if this correlation is valid. There could exist features outside of the dataset that could help to indicate if this correlation really represents the reality or if it is just a strong mathematical relationship.

```python
X_train_soybeans = soybeans.iloc[:, 1:13]
y_train_soybeans = soybeans.iloc[:, 13]
```

```python
sc = StandardScaler()  
X_train_soybeans = sc.fit_transform(X_train_soybeans)
```

## The Model

```python
regressor = xgboost.XGBRegressor(n_estimators=500000, learning_rate=0.08, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=7)
```

```python
regressor.fit(X_train_soybeans, y_train_soybeans)
```

```python
# feature importance
print(regressor.feature_importances_)
# plot
plt.style.use('ggplot')
plot_importance(regressor, color='red')
pyplot.show()
```

Feature importance didn't improve significatively the performance of the model. Using only
the feature 'price_soybeans' (as the plot suggests), I got RMSE = 143778.57 and a very bad accuracy, equals to -178.8165298211333.

```python
X_train_soybeans = soybeans.iloc[:, 1:13]
y_train_soybeans = soybeans.iloc[:, 13]
X_train_soybeans.drop(['gdp_thailand', 'gdp_world', 'gdp_egypt', 'gpd_netherlands',
                      'gdp_spain', 'gdp_iran'], axis=1, inplace=True)
X_test_soybeans = soybeans.iloc[L+1:l, 1:13]
X_test_soybeans.drop(['gdp_thailand', 'gdp_world', 'gdp_egypt', 'gpd_netherlands',
                      'gdp_spain', 'gdp_iran'], axis=1, inplace=True)
y_test_soybeans = soybeans.iloc[L+1:l, 13]
```

```python
X_train_soybeans = sc.fit_transform(X_train_soybeans)
X_test_soybeans = sc.fit_transform(X_test_soybeans)
```

```python
regressor.fit(X_train_soybeans, y_train_soybeans)
```

```python
predictions = regressor.predict(X_test_soybeans)
```

```python
print("RMSE: %.2f"
      % math.sqrt(np.mean((predictions - y_test_soybeans) ** 2)))

print(sklearn.metrics.explained_variance_score(predictions,y_test_soybeans))
```

```python
covariates2 = copy.deepcopy(df)
covariates2 = pd.DataFrame(df)
covariates2['year'] = covariates2['year'].astype(int)
covariates2 = covariates2[covariates2['year'] >= 2020]
covariates2.rename(columns={'year':'date'}, inplace=True)
#covariates.set_index('date', inplace=True)
#merged = pd.merge(covariates, comex, on='date')
#merged.drop(['state', 'route', 'usd', 'country'], axis=1, inplace=True)
#soybeans2 = covariates2[covariates['product'] == 'soybeans']
#soybeans.drop(['product', 'type'], axis=1, inplace=True)
X_answer_soybeans = covariates2.iloc[:, 1:13]
X_answer_soybeans.drop(['gdp_thailand', 'gdp_world', 'gdp_egypt', 'gpd_netherlands',
                      'gdp_spain', 'gdp_iran'], axis=1, inplace=True)
X_answer_soybeans = sc.fit_transform(X_answer_soybeans)
y_answer_soybeans = regressor.predict(X_answer_soybeans)
```
